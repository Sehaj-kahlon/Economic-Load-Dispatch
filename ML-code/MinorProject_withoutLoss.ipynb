```python
from sklearn.tree import DecisionTreeRegressor
# create datasets
X=df.iloc[:,4]
y=df[['P1','P2','P3','Total Cost']]
print(X)
```

```python
Xf=X.values.reshape(-1, 1)
```

```python
print(Xf)
```

```python
print(y)
```

```python
# define model
model = DecisionTreeRegressor()

# fit model
model.fit(Xf, y)

# make a prediction
row=[30]
yhat = model.predict([row])

# summarize prediction
print(yhat)
```

```python
print(yhat)
```

```python
p1=yhat[0][0]
p2=yhat[0][1]
p3=yhat[0][2]
total_cost=yhat[0][3]
ptotal=p1+p2+p3
print("P1=",p1,"\nP2=",p2,"\nP3=",p3,"\nTotal Power=",ptotal,"\nTotal Cost=",total_cost)
```
