```python
from sklearn.tree import DecisionTreeRegressor
# create datasets
X=df.iloc[:,8]
y=df.iloc[:,[0,1,2,3,4,5,7,9]]
print(X)
```

```python
Xf=X.values.reshape(-1, 1)
```

```python
print(Xf)
```

```python
print(y)
```

```python
# define model
model = DecisionTreeRegressor()

# fit model
model.fit(Xf, y)

# make a prediction
row=[900]
yhat = model.predict([row])

# summarize prediction
print(yhat)
```

```python
print(yhat)
```
