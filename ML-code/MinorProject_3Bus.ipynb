```python
from sklearn.tree import DecisionTreeRegressor
# create datasets
#Power Demand
X=df.iloc[:,5]
#P1, P2, P3, Power Loss , Totla Cost
y=df.iloc[:,[0,1,2,4,6]]
print(X)
```

```python
Xf=X.values.reshape(-1, 1)
#1D - 2D conversion
```

```python
print(Xf)
```

```python
print(y)
```

```python
# define model
model = DecisionTreeRegressor()

# fit model- train model
model.fit(Xf, y)

# make a prediction
row=[230]
yhat = model.predict([row])

# summarize prediction
print(yhat)
```

```python
print(yhat)
```

```python
p1=yhat[0][0]
p2=yhat[0][1]
p3=yhat[0][2]
cost=yhat[0][4]
ptotal=p1+p2+p3
ploss=yhat[0][3]
print("P1=",p1,"\nP2=",p2,"\nP3=",p3,"\nTotal Power=",ptotal,"\nPower Loss=",ploss,"\nTotal Cost= ",cost)
```
